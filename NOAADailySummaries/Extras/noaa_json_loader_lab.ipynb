# NOAA JSON Loader Lab

This notebook uses the separate `noaa_json_loader.py` module to load NOAA JSON files into DataFrames.

```python
from noaa_json_loader import load_noaa_json_dataframes
loaded = load_noaa_json_dataframes()
print('Source directory:', loaded.source_dir)
print('Files loaded:', len(loaded.file_dataframes))
print('Combined rows:', len(loaded.combined_dataframe))
```

```python
for filename, frame in loaded.file_dataframes.items():
    print(f'{filename}: rows={len(frame)}, cols={list(frame.columns)}')
```
